```python
from gensim.models import Word2Vec

print('loading w2v')
model = Word2Vec.load('FT_100/ft_oa_all_100d.bin')
```

```python
import numpy as np
import re

def embed(sent):
    tokens = sent.split()
    embeddings = []
    for token in tokens:
        try:
            embedding = model.wv.get_vector(token)
            embeddings.append(embedding)
        except KeyError:
            splits = re.sub(r'[^a-zA-Z]', ' ', token).split()
            for split in splits:
                try:
                    embedding = model.wv.get_vector(split)
                    embeddings.append(embedding)
                except KeyError: # 진짜 임베딩이 없는 단어인 경우
                    print(split)
    if not len(embeddings): # 임베딩된 토큰이 없을 경우 무의미한 단어로 채움
        embeddings.append(model.wv.get_vector('the'))
    embeddings = np.array(embeddings)

    return np.mean(embeddings, axis=0)
```
