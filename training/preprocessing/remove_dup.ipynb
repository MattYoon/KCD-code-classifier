```python
import pandas as pd

df = pd.read_csv('data_koeng_splited_translated.csv')
print(df.shape[0])
df = df[df['진단소분류'].notna()]
df = df[df['진단중분류'].notna()]
print(df.shape[0])
df.head(5)
```

```python
key_str = '진단코드'
val_str = '진단명'
dic = {}
count = 0
def check_diff(row, key_str, val_str):
    global count
    key, value = row[key_str], row[val_str]
    try:
        if dic[key] != value:
            count += 1
    except KeyError:
        dic[key] = value
df.apply(lambda row: check_diff(row, key_str, val_str), axis=1)
print('Done', count)
```

```python
dic = {}
def remove_diff(row, key_str, val_str):
    key, value = row[key_str], row[val_str]
    try:
        if dic[key] != value:
            row[val_str] = dic[key]
    except KeyError:
        dic[key] = value
    return row
df.apply(lambda row: remove_diff(row, key_str, val_str), axis=1)
print('Done')
```
