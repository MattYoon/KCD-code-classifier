```python
import pandas as pd
from tqdm import tqdm

tqdm.pandas()
df = pd.read_csv('data_final_encoded.csv')
df.head(10)
```

```python
df.agg(['count', 'size', 'nunique'])
```

```python
import re

junks = ['&nbsp;', '&lt;', '&gt;', 'gt','&amp;', '&quot;', '&apos;']
def clean_text(x):
    for junk in junks:
        x = x.replace(junk, '') # HTML 엔티티 및 탭 제거
        x = re.sub(r'(\n)\1+', r'\n', x) # 연속된 줄바꿈 제거
        x = x.replace(' \n', '\n') # 줄바꿈 전 공백 제거
        x = x.replace('\n', '. ') # 줄바꿈을 마침표로 변경
        x = x.replace('\t', ' ') # 줄바꿈을 마침표로 변경
        x = re.sub(r'[^a-zA-Z가-힣0-9 /+-,.]', ' ', x) # 영문, 한글, 숫자 및 특정 기호 외 제거
        x = x.replace(' .', '.') # 마침표 전 공백 제거
        x = re.sub(r'(\.)\1+', r'. ', x) # 연속된 마침표 제거
        x = x.replace(',.', ',')
        x = ' '.join(x.split()) # 연속된 공백 제거
    return x
```

```python
df['주호소 및 현병력'] = df['주호소 및 현병력'].progress_apply(clean_text)
df.head(10)
```

```python
df = df[df['주호소 및 현병력'].map(lambda x: len(x.strip()) >= 2)]
```

```python
df.agg(['count', 'size', 'nunique'])
```

```python
df.to_csv('data_cleaned.csv', index=False)
```
